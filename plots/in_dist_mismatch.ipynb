```python
import pandas as pd
import os

%load_ext autoreload
%autoreload 2

proj_folder = '/home/shared/generative_models/diffusion_overfit/data'
```

### Train-val distribution mismatch

```python
datasets = ['in64', 'in512', 'cifar10', 'cifar100']

def dist_shift_baseline(dataset, verbose=False):
    data = pd.read_csv(os.path.join(f"{proj_folder}/fd_analysis/train-v-val/{dataset}.csv"))

    fids, fdds = data['FID'], data['FDD']
    fid_mean, fid_std = fids.mean(), fids.std()
    fdd_mean, fdd_std = fdds.mean(), fdds.std()
    if verbose:
        print(f"On {dataset}, the distribution mismtach is \nFDD: {fdd_mean:.1f} +- {fdd_std:.1f} \nFID: {fid_mean:.2f} +- {fid_std:.2f}")
    return fdd_mean, fdd_std, fid_mean, fid_std

for i, dataset in enumerate(datasets):
    dist_shift_baseline(dataset, verbose=True)
```

### Train-train sampling noise

```python
datasets = ['in64', 'in512']

def sampling_baseline(dataset, verbose=False):
    data = pd.read_csv(os.path.join(f"{proj_folder}/fd_analysis/train-v-train/{dataset}.csv"))

    fids, fdds = data['FID'], data['FDD']
    fid_mean, fid_std = fids.mean(), fids.std()
    fdd_mean, fdd_std = fdds.mean(), fdds.std()
    if verbose:
        print(f"On {dataset}, the sampling noise is \nFDD: {fdd_mean:.1f} +- {fdd_std:.1f} \nFID: {fid_mean:.2f} +- {fid_std:.2f}")
    return fdd_mean, fdd_std, fid_mean, fid_std

for i, dataset in enumerate(datasets):
    sampling_baseline(dataset, verbose=True)
```
